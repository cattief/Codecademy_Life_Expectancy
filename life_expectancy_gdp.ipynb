```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('all_data.csv')
df = df.rename({'Life expectancy at birth (years)': 'life_exp'}, axis='columns')
```

```python
df.head(20)
```

```python
df.groupby(['Country']).count()
```

```python
df_chile = df[df.Country=='Chile']
df_cnr = df[df.Country=='China']
df_ger = df[df.Country=='Germany']
df_mexico = df[df.Country=='Mexico']
df_usa = df[df.Country=='United States of America']
df_zimb = df[df.Country=='Zimbabwe']
```

```python
df_chile.head()
```

```python
plt.figure(figsize=[16,9])
plt.plot(df_chile.Year,df_chile.life_exp)
plt.plot(df_cnr.Year,df_cnr.life_exp)
plt.plot(df_ger.Year,df_ger.life_exp)
plt.plot(df_mexico.Year,df_mexico.life_exp)
plt.plot(df_usa.Year,df_usa.life_exp)
plt.plot(df_zimb.Year,df_zimb.life_exp)
plt.legend(['Chile','China','Germany','Mexico','USA','Zimbabwe'])
plt.xlabel('Year')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy in diffenent countries')
plt.show()
```

```python
plt.figure(figsize=[16,9])
plt.plot(df_chile.Year,df_chile.GDP)
plt.plot(df_cnr.Year,df_cnr.GDP)
plt.plot(df_ger.Year,df_ger.GDP)
plt.plot(df_mexico.Year,df_mexico.GDP)
plt.plot(df_usa.Year,df_usa.GDP)
plt.plot(df_zimb.Year,df_zimb.GDP)
plt.legend(['Chile','China','Germany','Mexico','USA','Zimbabwe'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.title('GDP in diffenent countries')
plt.show()
```

```python
mean_df = df.groupby('Country',as_index=False).mean().drop('Year',axis=1)
```

```python
mean_df.head()
```

```python
countries = ['Chile','China','Germany','Mexico','USA','Zimbabwe']
ax = plt.subplot()
plt.pie(mean_df.life_exp)

total = mean_df.life_exp.sum()
print(total)

ax.pie(mean_df.life_exp,labels = countries, autopct=(lambda p: '{:.2f}'.format(p * total / 100)),
        shadow=True, startangle=90)
plt.show() 
```

```python
countries = ['Chile','China','Germany','Mexico','USA','Zimbabwe']
ax = plt.subplot()
plt.pie(mean_df.GDP)
total = mean_df.GDP.sum()
print(total)

ax.pie(mean_df.GDP,labels = countries, autopct=(lambda p: '{:e}'.format(p * total / 100)),
        shadow=True, startangle=0)
plt.show() 
```
